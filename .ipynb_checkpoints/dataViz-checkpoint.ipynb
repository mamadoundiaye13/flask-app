Lire des tableaux HTML avec Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame

langa_PROG = pd.read_html('https://fr.wikipedia.org/wiki/Chronologie_des_langages_de_programmation#')

df = pd.DataFrame()

```

```python
for i, val in enumerate(langa_PROG):
    df = pd.concat([df, DataFrame(langa_PROG[i],columns=['Années','Langage'])])
df = df.drop(df.index[0])
df.groupby(['Années'])
df
```
